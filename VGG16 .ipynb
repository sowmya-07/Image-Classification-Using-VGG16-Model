```python
#loading, preprocessing and converting to array
def preprocess(im):
    im=load_img(i,target_size=(224,224))
    im=img_to_array(im)
    im=np.expand_dims(im,axis=0)
    im=preprocess_input(im)
    return im
```

```python
pixels=[]
for i in os.listdir(DATA_DIR):
    #images=pd.Series(images,index=None).sample(frac=0.5).tolist()
    pixels.append(preprocess(i))
```

```python
#arrays of pixels
pixels
```

```python
from keras.applications.vgg16 import VGG16, preprocess_input, decode_predictions
```

```python
#VGG16 pretrained model
model=VGG16()
```

```python
#summary for model
model.summary()
```

```python
#predicting the images using VGG16 model 
pred = []
for i in pixels:
    pre = model.predict(i)
    pre = decode_predictions(pre,top=1)
    if pre is not None:
        pred.append(pre[0][0][1])
```
